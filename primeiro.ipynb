```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
vendas = pd.read_csv('20190420_VENDA.csv', encoding = "ISO-8859-1")
```

```python
def categoriza(hora):
    if hora <= '07:59':
        return '07:59h'
    elif hora >= '08:00' and hora <= '08:59':
        return '08:00h-08:59h'
    elif hora >= '09:00'and hora <= '09:59':
        return '09:00h-09:59h'
    elif hora >= '10:00'and hora <= '10:59':
        return '10:00h-10:59h'
    elif hora >= '11:00'and hora <= '11:59':
        return '11:00h-11:59h'
    elif hora >= '12:00'and hora <= '12:59':
        return '12:00h-12:59h'
    elif hora >= '13:00'and hora <= '13:59':
        return '13:00h-13:59h'
    elif hora >= '14:00'and hora <= '14:59':
        return '14:00h-14:59h'
    elif hora >= '15:00'and hora <= '15:59':
        return '15:00h-15:59h'
    elif hora >= '16:00'and hora <= '16:59':
        return '16:00h-16:59h'
    elif hora >= '17:00'and hora <= '17:59':
        return '17:00h-17:59h'
    elif hora >= '18:00'and hora <= '18:59':
        return '18:00h-18:59h'
    elif hora >= '19:00':
        return '19h'
    
vendas['FAIXA_HORARIO'] = vendas['HORA'].apply(categoriza)
```

```python
def categoriza(TOTAL):
    if TOTAL <= 9.99:
        return 9
    elif TOTAL >= 10 and TOTAL <= 19.99:
        return 19
    elif TOTAL >= 20 and TOTAL <= 29.99:
        return 29
    elif TOTAL >= 30 and TOTAL <= 39.99:
        return 39
    elif TOTAL >= 40 and TOTAL <= 49.99:
        return 49
    elif TOTAL >= 50 and TOTAL <= 59.99:
        return 59
    elif TOTAL >= 60 and TOTAL <= 69.99:
        return 69
    elif TOTAL >= 70 and TOTAL <= 79.99:
        return 79
    elif TOTAL >= 80 and TOTAL <= 89.99:
        return 89
    elif TOTAL >= 90 and TOTAL <= 99.99:
        return 99
    elif TOTAL >= 100:
        return 100
    
vendas['FAIXA_VALOR'] = vendas['TOTAL'].apply(categoriza)
```
